```python
# Eliminar filas donde 'Fecha compra' es nulo
df_ventas = df_ventas.dropna(subset=["Fecha compra"])
```

```python

ventas_ayores  = df_ventas[df_ventas["Cantidad"] > 13]

ventas_ayores.count()
```

```python
cantidades = df_ventas[df_ventas["Cantidad"] > 13].index
df_ventas = df_ventas.drop(cantidades)
df_ventas
```

Completo las fechas de envio que se encuentran vacias 

```python
df_ventas["Fecha compra"] = pd.to_datetime(df_ventas["Fecha compra"])
df_ventas["Fecha envío"] = pd.to_datetime(df_ventas["Fecha envío"])
```

```python
df_ventas["Fecha envío"].dtype
```

```python
fechas_resto = df_ventas["Fecha envío"] - df_ventas["Fecha compra"]
fechas_resto = fechas_resto.dt.days
medias_dias = fechas_resto.mean()

```

```python
def Reemplazo_NA(x):
    if pd.isna(x):
        return df_ventas["Fecha compra"].iloc[0] + pd.to_timedelta(medias_dias,unit='D')
    return x 
```

```python
df_ventas["Fecha envío"] = df_ventas["Fecha envío"].apply(Reemplazo_NA)

print(df_ventas["Fecha envío"])
```

transformo en INT 

```python
df_ventas["Precio unitario"] = df_ventas["Precio unitario"].str.replace(',','')
df_ventas["Precio unitario"] = pd.to_numeric(df_ventas["Precio unitario"],errors = 'coerce')
```

```python
df_Productos["Coste Producción"] = pd.to_numeric(df_Productos["Coste Producción"],errors = 'coerce')
```

```python
m1 = df_clientes.merge(df_ventas, on="ID Cliente",how="inner")
m1
```

```python
df_global = df_Productos.merge(m1, on="ID Producto",how="inner")
df_global
```

```python
df_global["Volumen_Venta"] = df_global["Cantidad"] * df_global["Precio unitario"]
df_global["Año_compra"] = df_global["Fecha compra"].dt.year
df_global["Beneficio"] = df_global["Volumen_Venta"] - df_global["Coste Producción"]
```

```python
url = "https://en.wikipedia.org/wiki/List_of_countries_by_average_wage"
r = requests.get(url)
df_list = pd.read_html(r.text)
df=df_list[1]
df
```

```python
df_renta=df.rename(columns={"Country":"País",	"Gross Average Monthly Wages (USD, at current exchange rates)[4]":"Sueldo Mensual USD RPC",	"Year":"Año"})
```

```python
df_renta= df_renta.drop(columns=["Año"])
```

```python
df_renta["País"] = df_renta["País"].str.replace(r'[$,*]','',regex = True)
df_renta["País"] = df_renta["País"].str.strip()
df_global["País"] = df_global["País"].str.strip()
```

```python
##se agrega el sueldo mensual de renta per capita en caso de querer buscar correlaciones 
df_global = df_global.merge(df_renta, on= "País",how="left")
df_global
```
